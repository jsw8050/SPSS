1. 우발적 사건: 조사 설계 이전 또는 진행 중 예기치 못하거나 일시적으로 발생한 상황이 타당도 해침

1. 개념: 모집단을 비슷한 성질을 가진 2개 이상의 **동질적인** 층으로 나눈 후, 각 층에서 단순무작위추출법을 이용하여 표본 추출하는 방법

2. 장점

    - 표본 크기가 크지 않아도 모집단의 대표성 보장
    
3. 단점
    - 표본 추출과정에서 시간 많이 소요
    - 모집단의 각 층에 대한 정확한 정보 필요
    - 모집단을 층화하여 가중하였을 경우 원형 복귀 어렵

### 9. 횡단적 조사, 종단적 조사
- 횡단적 조사는 20대, 30대, 40대로 구성된 모집단에서 모집단의 크기에 비례하여 무작위 표본을 추출한 후 일정 시점에서 서로 다른 특성을 가진 표본 집단을 한꺼번에 조사하는 표본 조사, 한 시점에서 연령대별 특정 제품에 대한 선호도에 차이가 있는지 파악
- 종단적 조사는 표본 집단을 일정한 시간적 간격을 두고 시간의 변화에 따른 특정 제품에 대한 선호도를 반복적으로 조사하여 선호도의 변화 상황을 파악한다. 20대로 구성된 모집단에서 무작위로 표본을 추출한 후 특정제품에 대한 선호도를 조사한 후, 해당 표본이 시간이 흐름에 따라 연령대가 변화하면서 특정제품에 대한 선호도가 어떻게 변화하는지 조사

### 1. 서스톤 척도의 개발 과정을 3단계로 구분하여 설명하시오.
- 연구하고자 하는 태도와 관련된 문항을 수십 또는 수백 개 정도 만든 후, 다수의 평가자들로 하여금 이들 문항을 소수의 범주로 나누게 한다.
- 평가자 개개인들이 판단하여 위치시킨 점을 각 문항의 점수로 하여 평가자들의 수로 나눈 척도값을 계산하고, 점수분포가 지나치게 분산된 문항들은 원칙적으로 태도가 불분명한 것으로 간주하여 채택하지 않는다.
- 남은 문항들 중에서 태도의 한 극단에서 다른 극단에 고르게 분포되도록 문항들을 선정하여 척도를 구성한다.

3. 등간척도: 측정 대상의 속성에 따라 서열화할 뿐 아니라 서열 간의 간격에 있어 동일성 유지하도록 수치 부여, 온도, 지능지수
4. 비율척도: 등간척도의 성격에 절대영점이 존재하는 척도를 가지고 수치를 부여한 것, 키, 몸무게

1. 층화표본추출법: 모집단을 비슷한 성질을 가진 2개 이상의 동질적인 층으로 나누어 각 층에서 단순무작위추출방법을 사용하여 표본 추출
    - 층 내는 동질적, 층 간은 이질적
    - 표본추출 단위는 구성요소
2. 집락표본추출법: 모집단을 집계 단위, 조사단위로 모은 후 군집으로 나누고, 이들 군집들 중에서 일부를 추출한 후, 추출된 군집 중에서 일부 또는 전부를 표본으로 추출
    - 집락 간 동질적, 집락 내 이질적
    - 표본추출 단위는 집락
    - 집락 내부가 모집단이 지닌특성의 분포와 정확히 일치하면 가장 이상적이다.

### 8. 표본오류와 비표본 오류와 그 원인 
1. 표본오류: 모집단으로부터 표본을 추출하여 조사한 자료를 근거로 얻은 결과를 모집단 전체에 대해 일반화하기 때문에 필연적으로 발생하는 오류

2. 비표본오류: 표본오류를 제외한 모든 오류, 면접이나 조사표 작성 등에서 나타나는 오류
 
3. 원인
    - 조사자의 편견
    - 입력 오류
    - 무응답
    - 조사단위 누락 등
    - 조사 착오
