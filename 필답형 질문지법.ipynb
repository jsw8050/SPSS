### 2-1. 등간척도 항목의 문제점
1. 만족도 점수가 등간척도로 되어 있어 수학적으로 >, <, = ,!=, +, -는 가능하지만 *, / 가 불가능하다.
2. 절대영점이 존재하는 비율 척도로 구성하여 수학적으로 *, / 까지 가능할 수 있게 있음에도 불구하고 등간척도로 구성하고 많은 정보를 상실할 수 있다.

### 2-2. 위의 질문형태 택한 이유
등간 척도를 택함으로써 측정 대상의 속성에 따라 서열화하는 것뿐만 아니라 서열 간 동일한 간격을 유지하도록 수치를 부여할 수 있다. 수학적으로 +, -가 가능하다.

### 2-3. 문제점 완화 방법
등간 척도의 성격에 더불어 절대영점이 존재하는 비율척도를 만들어 모든 수학적 연산이 가능하도록 한다.

단, 만족도가 없으면 '0'으로 기입해주십시오.
