## 3회
### 1. 비확률표본표출방법과 확률표본추출방법의 특징
1. 확률표본표출방법
    - 무작위적 표본추출
    - 모수 추정 편의 없음
    - 표본 오차 측정 가능
    - 분석 결과에 대한 일반화 가능
    - 시간 비용이 많이 듦
    - 연구 대상이 표본으로 추출될 확률 알려져 있음
    
2. 비확률표본추출방법
    - 작위적 표본추출
    - 모수추정에 편의 있음
    - 시간과 비용 적게 듦
    - 연구 대상이 표본으로 추출될 확률이 알려져있지 않음
    - 표본 오차 측정 불가능
    - 분석 결과 일반화 제약


### 2. 변수 간의 선형관계, 대칭관계, 인과 관계 개념을 쓰시오.
1. 선형관계
    - 변수들 간의 직선적 관계, ex) 표본 상관계수

2. 대칭관계
    - 두 관찰자 간의 측정 범주 값에 대한 일치도 측정 ex) 교차 분석의 대칭적 측도를 나타내는 카파, 맥네마 
    
3. 인과 관계
    - 원인이 되는 변수인 독립변수가 영향을 받는 변수인 종속변수에 미치는 관계 ex) 회귀분석

### 4. 코호트연구와 패널 연구의 공통점, 차이점
- 공통점: 한 연구 대상을 일정 기간 동안 관찰하여 그 대상의 변화를 파악하는 데에 초점을 두는 종단 조사의 방법 
- 차이점: 패널 연구는 연구 대상이 고정되어 있는 반면, 코호트연구는 동일한 시기에 동일한 경험을 한 대상의 특징을 갖는다.


### 5. 바람직한 실험설계 조건을 쓰시오.
실 외 독 종
- 실험 대상의 무작위화: 연구 대상을 실험변수와 통제변수로 나눌 때 가능한 두 변수 간 차이가 적도록 무작위 할당
- 외생 변수의 통제: 외생변수는 종속변수에 영향을 미칠 수 있는 변수, 인과 관계 파악하는 데에 문제 발생
- 독립변수의 조작: 연구자가 의도적으로 어떤 한 집단에는 독립변수를 발생시키고, 다른 집단에는 발생하지 않도록 한 후 독립변수의 조작이 실험변수에 미치는 영향을 관찰
- 종속변수의 비교: 실험집단과 통제집단 간의 종속변수를 비교하거나, 실험 전후의 종속변수를 비교하여 두 변수 간의 차이가 있는지 확인

### 6. 확률표집 방법
단 층 계 집
1. 단순임의추출법: 적절한 표본 크기를 결정하여, 표본이 추출될 확률이 동일하도록 랜덤으로 표본크기만큼 표본 추출
    - 장점: 사전 지식 필요 없음, 표본 오차 계산 용이
    - 단점: 모집단 사전 지식 활용 불가, 표본 규모가 커야 함

2. 층화추출법: 모집단을 비슷한 성질을 가진 2개 이상의 동질적인 층으로 나눈 뒤, 각 층에서 단순무작위추출법을 통해 표본 추출  
    - 장점: 표본이 크지 않아도 대표성 보장, 불필요한 자료 분산 최소화
    - 단점: 비용, 시간 많이 듦, 모집단 각 층에 대한 정확한 정보 필요

3. 계통추출법: 모집단 크기가 N이고 표본 크기가 n일 때, 난수표를 이용하여 최초의 표본만 무작위로 추출하고, 나머지는 일정한 간격(k)을 두고 표본 크기만큼 표본 추출
    - 장점: 단순무작위 추출법 보다는 대표성 존재, 선정 쉽고, 비용 적음
    - 단점: 표본이 주기성 있는 경우 오류, 표본 나열 시 연구자의 편견 개입 가능
    
4. 집락추출법: 모집단을 조사 단위 또는 집계 단위로 모은 후 단순임의추출법으로 군집을 추출하여 군집으로 구분하고, 이들 군집 중 일부의 군집을 추출한 뒤 추출된 군집에서 일부 또는 전부를 표본으로 추출하는 방법
    - 장점: 모집단 목록 없는 경우 유리, 비용 절감
    - 단점: 집단 내 이질성 확보 어렵, 표본 크기가 커야함

### 7. 개념타당도 
수 판 이

- 수렴타당도: 동일한 개념을 측정하기 위해서는 서로 다른 측정 방법을 통해 얻은 측정치들의 상관관계가 높아야 함을 전제로 하는 타당도
- 판별타당도: 서로 다른 개념을 측정했을 때 각 측정 문항들의 결과 간의 상관관계가 낮아야 함을 전제로 하는 타당도
- 이해타당도: 특정 개념에 대해 이론적 구성을 토대로 논리적, 체계적으로 어느 정도 이해하고 있는가를 나타내는 타당도

### 8. 개방형 질문과 폐쇄형 질문의 장단점
1. 개방형 질문: 응답자가 자유롭게 응답할 수 있는 질문
    - 장점: 복합적인 질문을 할 때 유리, 새로운 사실 발견 가능, 응답유형에 대한 사전 지식이 없을 경우 용이
    - 단점: 응답자들이 어느 정도의 교육 수준을 갖춰야 한다. 응답 결과 분류 및 코딩 어려움, 응답률 낮음, 결과 분석 후 설문지 제작에 시간 오래 걸림
    
2. 폐쇄형 질문: 사전에 연구자가 선택지를 제시한 뒤 택하게 하는 질문  (적변의 편)
    - 장점: 코딩 용이, 응답 관련 오류 적음, 사적인 질문 시 용이, 연구자의 편견 개입 방지
    - 단점: 응답지가 획일화되어 있어서 응답자들의 의견과 달라 편향 발생 가능, 적절한 응답지 제시 어렵, 응답자 의견 충분히 반영 어렵,질문의 순서 변화 시 응답 내용에 변화 발생
