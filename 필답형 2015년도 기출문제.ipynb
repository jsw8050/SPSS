## 1회
#### 1. 단일집단 사후실험설계의 특징을 간략히 설명하고 장단점을 쓰시오.

- 통제집단을 따로 두지 않고 실험집단에만 실험 처리를 한 후 어느 정도 시간이 지난 후에 이 실험의 효과를 측정하는 방법으로 어떤 실험으로 인해 야기되는 구체적인 변화를 분석하기 보다는 탐색적 목적을 위해 사용하는 실험 설계이다.
- 장점: 실험 설계가 용이하다. / 두 변인 간의 단순 관계 파악이 용이하다. / 독립 변수 조작에 따른 변화의 관찰이 제한된 경우 사용 가능하다.
- 단점: 외적, 내적 타당도가 낮다. / 외생변수의 통제가 불가능하다. / 분석 결과를 일반화할 수 없다.

#### 2. 집단 조사의 의미를 설명하고 장단점을 쓰시오.

- 연구 대상을 개별적으로 만나는 것이 아니라 집단적으로 모아놓고 설문지를 배부하여 대상자들이 직접 기입하게 하는 방식
- 장점: 시간과 비용을 절감한다. / 조사의 설명이나 조건을 똑같이 할 수 있어서 동질성 확보 / 필요 시 응답자와 직접 대화할 수 있기 때문에 질문에 대한 오류를 줄일 수 있다.
- 단점: 응답자 집합시키는 것이 어렵기 때문에 특수한 조사에서만 가능하다. / 응답자들의 개인차를 무시하기 때문에 조사 자체에 타당도가 낮아진다. / 응답자들이 한 곳에 모여 있기 때문에 통제가 용이하지 않다.


#### 3. 신뢰성을 향상시키기 위한 방법 5가지

- 개념: 측정 도구가 측정하고자 하는 현상을 일관성있게 측정하는 능력으로, 어떤 측정 도구를 동일한 현상에 반복 적용하여 동일한 결과를 얻게 되는 정도를 말한다.
- 방법 (다모일 대표측)
    - 다수의 측정 항목 
    - 측정 도구의 모호성 제거
    - 측정의 일관성 유지
    - 대조적인 항목 비교
    - 표준화된 측정 도구 이용
    - 측정자가 무관심하거나 잘 모르는 내용은 측정하지 않는다.

#### 4. 변수의 개념적 정의와 조작적 정의를 예를 들어 설명하시오.
- 개념적 정의: 하나의 개념을 정의하기 위해 다른 개념을 이용하여 묘사하는 것이다.
- 조작적 정의: 추상적인 개념들은 경험적, 실증적으로 **측정 가능하도록 구체화**한 것이다.
- 예: 지능은 개념적 정의에서 문제해결 능력이라고 정의할 수 있고, 조작적 정의에서는 측정이 가능한 IQ 검사에 의한 점수로 정의할 수 있다.

#### 5. 단순무작위표본추출법을 실제조사에 활용하기 쉽지 않은 이유 세 가지

(노오크)
- 현실적으로 많은 노력이 필요하다.
- 동일한 표본크기에서 층화추출법보다 표본오차가 크다.
- 비교적 표본의 크기가 커야 한다.

#### 6. 체계적 오차와 비체계적 오차의 의미를 쓰시오.

- 체계적 오차: 편향이라고도 하며 어떠한 영향이 측정대상에 체계적으로 미침으로써 일정한 방향성을 갖는 오차로 측정의 타당성과 관련이 있다. 

- 비체계적 오차: 측정 과정에서 우연히 또는 일시적으로 발생하는 불규칙적인 오차로 측정의 신뢰성과 관련이 있다.

#### 7. 확률 표집과 비확률표집의 특징을 비교 설명하시오. 비확률표집 네 가지를 쓰고 각각 설명하시오. 

- 확률표본추출방법
    - 무작위 표본 추출
    - 모수 추정에 편의가 없음
    - 분석 결과 일반화 가능
    - 시간과 비용이 많이 듦
    - 연구 대상이 표본으로 추출될 확률이 알려져 있음
    - 표본오차 측정 가능
- 비확률표본추출방법
    - 시간과 비용이 적게 듦
    - 모수 추정에 편의가 있음
    - 표본 오차 측정 불가능
    - 연구 대상이 표본으로 추출될 확률이 알려지지 않음
    - 분석 결과 일반화 불가능
    - 작위적 표본 추출
- 비확률표집 방법
    - 눈덩이 추출: 눈덩이를 굴리면 커지는 것처럼 소수의 응답자를 찾은 다음 그와 비슷한 사람들을 소개받아 가는 방식으로 표본 추출
    - 판단 추출: 조사원이 자신의 경험과 지식에 의해 모집단을 가장 잘 대표한다고 여겨지는 표본을 주관적으로 판단하여 표본을 추출
    - 편의 추출: 모집단에 대한 정보가 없거나 모집단 구성요소 간 차이가 없는 경우 편리성에 기준을 두고 조사원이 마음대로 표본을 선정하는 방법
    - 할당 추출: 모집단이 여러가지 특성으로 구성되어 있는 경우 각 특성에 따라 층을 구성한 다음, **층별 크기에 비례하여 표본을 배분하거나 조사원이 그 층 내에서 직접 선정**하여 조사하는 방법

#### 8. 조사 연구의 결과를 해석할 때 분석단위에 대한 잘못된 이해에서 야기될 수 있는 오류를 쓰시오

1. 생태학적 오류
    - 분석단위를 집단으로 두고 얻어진 결과를 개인에게 적용하면서 생기는 오류이다. 예를 들어 흑인이 많이 사는 도시의 범죄율이 높다고 실제로 흑인들이 범죄를 저질렀다고 판단하면 안되다.
2. 개인주의적 오류
    - 분석 단위를 개인으로 두고 얻어진 결과를 집단에 적용하면서 생기는 오류이다. 예를 들어 흑인들의 범죄율이 높다고해서 실제로 흑인들이 많이 사는 도시의 범죄율이 높다고 판다하면 안된다.

#### 9. 층화표집과 집락표집의 기본개념을 설명하시오. 
1. 층화표집: 모집단을 비슷한 성질을 갖는 2개 이상의 동질적인 층으로 구분하고, 각 층으로부터 단순무작위추출방법을 적용하여 표본을 추출하는 방법 
    - 집단 내 차이: 동질적
    - 집단 간 차이: 이질적
2. 집락표집: 모집단을 조사단위 또는 집계단위로 모은 군집으로 구분하고, 이들 군집 중 일부의 군집을 추출한 뒤 추출된 군집에서 일부 또는 전부를 표본으로 추출하는 방법
    - 집단 내: 이질적
    - 집단 간: 동질적

## 2회
#### 1. 실험의 내적, 외적 타당성을 저해하는 외생변수 종류 세 가지를 쓰고 설명하시오.
- 외생변수는 종속 변수에 영향을 미치는 독립 변수를 제외한 모든 변수. 외생변수를 억제하지 못하면 실험 변수와 결과 변수 간의 인과 관계를 파악하기 어렵다.
- 통제 방법: 상쇄, 제거, 무작위화
- 종류
    - 우발적 사건: 조사 설계 이전 또는 설계 과정에서 예기치 못하게 발생한 상황이 타당도를 해친다.
    - 성숙효과: 실험 기간 중에 실험 집단이 육체적 또는 심리적 특성이 변화함으로써 실험 결과에 영향을 미쳐 타당도를 해친다.
    - 사멸효과: 실험 대상의 일부가 사망 또는 사멸 등의 이유로 추적 불가능해지는 경우 실험결과에 영향을 미쳐 타당도를 해치게 된다.
    - 통계적 회귀: 사전 측정에서 극단적인 값을 얻은 경우 이를 여러 번 반복 측정하게 되면 평균치로 근사하게 되는 경향으로 타당도를 해치게 된다.
    - 조사도구 효고: 자료를 수집하는 과정에서 사용되는 도구가 달라지는 경우 측정결과에 영향을 미쳐 타당도를 해친다.


#### 2. 지수와 척도의 의미를 쓰고 사회과학에서 지수와 척도를 사용하는 이유 두 가지를 쓰시오.

1. 지수: 같은 종류의 통계수치에 대한 대소 관계를 비율의 형태로 표시한 것
    - 장소적 비교, 시간저 비교 모두에서 사용 가능
    - 변수의 변화에 대한 비교가 용이하다.
    - 서로 다른 단위로 측정한 변량의 집계 가능
2. 척도: 논리적으로 정해진 규척에 따라 대상의 속성에 값을 부여하는 것
    - 표준화된 척도를 일관되게 사용하면 측정의 신뢰도가 올라간다.
    - 변수에 대한 양적 관찰값을 제공하여 통게적 자료 분석이 가능하다.
    - 여러개의 지표를 하나의 점수로 요약하여 나타냄으로써 하나의 지표로 나타내기 어려운 복합적인 개념을 측정할 수 있다.

#### 3. 신뢰도를 높이기 위한 방안 네 가지

1. 신뢰도: 측정도구가 측정하고자 하는 현상을 일관성 있게 측정하는 능력
2. 방법
    - 측정도구의 모호성 제거
    - 측정자가 무관심하거나 잘 모르는 내용은 측정하지 않는다.
    - 다수의 측정 항목
    - 측정의 일관성 유지
    - 대조적인 항목들 비교
    - 표준화된 측정 도구 이용

#### 4. 우편조사와 비교하여 면접조사의 장단점 쓰기
1. 면접조사의 장점
    - 응답의 허위 여부 확인 가능
    - 우편조사에 비해 회수율 높다.
    - 보조도구 사용 가능
    

2. 단점 (조여시특)
    - 조사자에 따라 오차 발생 가능
    - 응답자가 생각할 여유가 없다.
    - 시간과 비용이 많이 든다.
    - 특수층의 사람 만나기 어렵다.

#### 5. 1차자료와 2차자료의 의미와 예시
1. 1차 자료
    - 연구자가 현재 수행 중인 연구 목적을 수행하기 위해서 직접 수집한 자료로, 면접법, 설문지법, 관찰법 등을 이용한다.
    
2. 2차 자료
    - 다른 목적을 위해 이미 수집된 자료로 연구자가 자신이 수행 중인 연구문제를 해결하기 위해 사용하는 자료
    
3. 예시
    - "외국인 근로자에 대한 직장 만족도 조사"를 하는 경우 우선 직장에 근무하는 외국인에 대한 자료를 직접 수집하는 대신 노동부로부터 직장에 근무하는 외국인에 대한 자료(2차 자료)를 제공받아 연구 목적과 내용에 맞게 조사 설문지(1차 자료)를 만들어 조사한다.

#### 6. 유사실험설계의 장단점
무작위할당에 의해 실험집단과 통제집단을 동등하게 할 수 없는 경우 무작위할당 대신 실험집단과 유사한 비교집단을 구성하여 실험하는 설계

- 장점: (일 이 복)
    - 실제 상황에서 이루어지므로 일반화 가능성 높다.
    - 일상 생활과 동일한 상황에서 수행되므로 이론적 검증 및 현실문제 해결에 유용하다.
    - 복잡한 사회적, 심리적 영향과 과정변화 연구에 적합하다.
    
- 단점: (무 정 외)
    - 현장상황에서는 대상의 무작위화와 독립변수의 조작화 어렵
    - 측정과 외생변수 통제가 어려워서 연구결과의 정밀도 떨어짐
    - 실제 상황이라서 독립변수 효과와 외생변수 효과 분리 어렵

#### 7. 집락표본추출 절차와 장단점
1. 집락표본추출: 모집단을 집계 단위 또는 조사 단위로 모아 집락으로 나누고 이들 중 일부의 군집을 추출한 다음 일부 또는 전부를 표본으로 추출하는 방법

2. 절차
    - 모집단을 여러개의 군집으로 형성
    - 단순임의표본추출법에 의하여 군집을 추출하여 군집표본 구성
    - 추출된 군집 내에 있는 모든 대상 또는 일부를 표본조사단위로 하여 표본 구성
    
3. 장점 ( 신 성 전 시 )
    - 시간 비용 절감
    - 전체 모집단 목록 작성 불필요
    - 각 집락 성격 및 모집단 성격 파악 가능
    - 신속한 조사결과 산출
    
4. 단점 (오 집 단)
    - 집락이 동질적이면 오차의 개입 가능성 높음
    - 오차의 개입 방지 또는 평가 방법 없음
    - 단순무작위표본추출보다 특정집단은 과대, 과소 표현할 위험 크다.

#### 8. 표본추출의 기본용어인 표집간격, 표집률, 모수, 통계량 의미
모집단의 크기를 N이라고 하고, 표본의 크기를 n이라고 할 때,
- 표집간격: N/n
- 표집률: n/N
- 모수: 모집단을 대표하는 수치화된 특정한 값
- 통계량: 표본의 관측값으로부터 계산되는 확률변수로서 표본을 택할 때마다 변하는 값

#### 9. 표본오류와 비표본 오류, 비표본오류 중 불포함 오류와 무응답오류에 대해 설명하시오.
- 표본오차: 모집단으로부터 표본을 추출하여 조사한 자료를 얻은 결과를 모집단 전체에 일반화하기 때문에 필연적으로 생기는 오차
- 비표본오차: 표본오차를 제외한 모든 오차
- 불포함오차: 표본체계의 불완전으로 모집단에는 속해있으나 표본집단에 선정될 수 없는 경우에 발생하는 오차
- 무응답오차: 표본추출 과정에서 선정된 표본 중 일부가 연결이 되지 않거나 응답을 거부당했을 때 발생하는 오차
