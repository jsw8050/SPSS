3-1. 등간 척도의 문제
    - 등간 척도는 수학적으로 <,>,=, !=, +, - 는 표현할 수 있으나 *, /는 계산할 수 없다.
    - 절대 영점을 포함하는 비례척도를 이용할 수 있음에도 불구하고 등간척도를 이용하여 더 많은 많은 정보를 얻을 수 있는 기회를 상실했다.
    
3-2.등간 척도 선택 이유
서열 척도는 측정 대상의 속성에 따라 각 측정 대상 간의 등급 순위를 지정하여 수치화한 것으로 >, <, =, != 는 가능하지만 등간 척도를 이용하면서열화 뿐만 아니라 +, - 계산이 가능하기 때문이다.
